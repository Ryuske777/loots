```python
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
import wikipedia as wp

result = wp.page("madhuri dixit") # accessing the wikipedia page
final_result = result.content
def plot_wordcloud(wc):
    plt.axis("off")
    plt.figure(figsize=(10,10))
    plt.imshow(wc)
    plt.show()
wc=WordCloud(width=500, height=500, background_color="deepskyblue", random_state=10, 
             stopwords=STOPWORDS).generate(final_result)
wc.to_file("ComSc.png")
plot_wordcloud(wc)


# this basically shows related words according to given input name and it also created a png file
```
