```python
#checked and working
import networkx as nx
import pylab as plt
G=nx.DiGraph()
G.add_weighted_edges_from([('A','B',1),('A','C',1),('C','A',1),('B','C',1)])
ppr1=nx.pagerank(G)
print("Page rank value:",ppr1)
pos=nx.spiral_layout(G)
nx.draw(G,pos,with_labels=True,node_color="#f86e00")
plt.show()
```
